#### Checking the most common words used in the plot  

```python
academyplot= pd.DataFrame(academy['plot'].dropna(axis=0, how='any'))
```

```python
academyplot.head()
```

```python
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
words = " ".join(academyplot['plot'])

def punctuation_stop(text):

    filtered = []
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    for w in word_tokens:
        if w not in stop_words and w.isalpha():
            filtered.append(w.lower())
    return filtered


words_filtered = punctuation_stop(words)

text = " ".join([ele for ele in words_filtered])

wc= WordCloud(background_color="white", random_state=1,stopwords=STOPWORDS, max_words = 2000, width =800, height = 1500)
wc.generate(text)

plt.figure(figsize=[10,10])
plt.imshow(wc,interpolation="bilinear")
plt.axis('off')
plt.show()
```

We can see that some of the most common words in the plot are: Young, Life, american, world war, family, love, woman, find.
